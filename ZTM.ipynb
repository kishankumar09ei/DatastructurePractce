```python
def find_two_sum(nums, target):

    pairs=[]
    pair_dct={}
    for i in range(len(nums)):
        complement=target-nums[i]
        if pair_dct.get(nums[i],None) is None:
            pair_dct[complement]=i
        else:
            pairs.append([i,pair_dct.get(complement,i)])
        
    return pairs

print(find_two_sum([1,2,7,9,2], 11))
print(find_two_sum([1,4,7,9,2], 11))
print(find_two_sum([1,3,7,9,2], 11))
print(find_two_sum([1,3,7,9,2], 26))
print(find_two_sum([], 0))
print(find_two_sum([5], 5))
print(find_two_sum([1,6], 7))
```

```python
# we need pair of numbers which distance is maximum and multiplication of min no in pair * distance should be maximum
#So we can take two pointers one from start and one from end,calulate distance and min*distance is max.That is our value

def maxArea(height) :
    max_area=0
    if len(height)>1:
        for p1 in range(len(height)):
            for p2 in range(p1+1,len(height)):
                if min(height[p1],height[p2])*(p2-p1) > max_area:
                    max_area=min(height[p1],height[p2])*(p2-p1)
                #print(p1,p2)
        return max_area
    else:
        return 0
    

print(maxArea([1,8,6,2,5,4,8,3,7]))
print(maxArea([1,1]))

print(maxArea([1,8,6,2,9,4]))
print(maxArea([1,7,2,0,1,3]))
print(maxArea([1,8,6,2,9,4]))
print(maxArea([]))
print(maxArea([2]))
```

```python
def getmaxArea(height) :
    max_area=0
    if len(height)>1:
        p1=0
        p2=len(height)-1
        while p1<p2:
            if min(height[p1],height[p2])*(p2-p1) > max_area:
                max_area=min(height[p1],height[p2])*(p2-p1)
            if height[p1]>height[p2]:
                p1+=1
            else:
                p2-=1
        return max_area
    else:
        return 0
    
       
print(maxArea([1,8,6,2,5,4,8,3,7]))
print(maxArea([1,1]))

print(maxArea([1,8,6,2,9,4]))
print(maxArea([1,7,2,0,1,3]))
print(maxArea([1,8,6,2,9,4]))
print(maxArea([]))
print(maxArea([2]))
```

```python
height = [0,1,0,2,1,0,1,3,2,1,2,1]

def get_max_water(height):
    total_water=0
    for i in range(len(height)):
        left_max=max(height[:i+1])
        right_max=max(height[i:])
        total_water+=min(left_max,right_max)-height[i]
    return total_water
get_max_water(height)
```

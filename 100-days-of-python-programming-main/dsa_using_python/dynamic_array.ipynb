### we will try to create a python list from array and create below functionality


```python
import ctypes
# to create C type ka array
class MyList:

    def __init__(self):
        self.size=1## current size 
        self.n=0 #current number of elements in array
        #create a c type array with size -> self.size
        self.lst= self.__make_array(self.size)

    def __make_array(self,capacity):
        return (capacity*ctypes.py_object)()

    def __len__(self):
        return self.n
    
    def append(self,value):
        ##check if it is vacant
        if self.size== self.n:
            ##rray is full : resize
            self.__resize(self.size*2)
        self.lst[self.n]=value
        self.n += 1

    def __resize(self,new_capacity):
        ##here we need to create a new array of 
        ## new_capacity and add element old array to new array
        ## ass

        ##create a new array 
        new_array= self.__make_array(new_capacity)
        ##copy old array elements to new array
        for i in range(self.n):
            new_array[i]=self.lst[i]
        self.size= new_capacity
        #replace variable lst to new_array 
        self.lst=new_array
    
    def __str__(self):
        result=''
        for i in range(self.n):
            result += str(self.lst[i]) +','
        return '['+ result[:-1]+']'
    
    def __getitem__(self, index):
        if index>=0:
            if index < self.n:
                return self.lst[index]
        elif -1*(self.n) <= index:
            return self.lst[self.n + index]
        else:
            return "Index out of range error"
        
    def pop(self):
        if self.n==0:
          return "Empty List"
        #get the last element
        print(self.lst[self.n-1])
        self.n -=1
        
    def clear(self):
        """ 
        remove all elements of list, kinda  reset
        """
        self.n=0
        self.size=1

    def find(self,value):
        for i in range(self.n):
            if self.lst[i]==value:
                return i
        return "value not found"
    
    def insert(self,value,pos):
        if self.n== self.size:
            self.__resize(self.size*2)
        for i in range(self.n,pos,-1):
            self.lst[i]=self.lst[i-1]
        
        self.lst[pos]=value
        self.n+=1

    def remove(self,value):
        ## find the item index
        pos=self.find(value)
        if type(pos) == int:
            self.__delitem__(pos)
        else:
            return pos

    def __delitem__(self, key):
        # delete pos wala item
        if 0<=key<self.n-1:
            for i in range(key,self.n-1):
                self.lst[i]=self.lst[i+1]
            self.n-=1
        else:
            return "index out of range"
```

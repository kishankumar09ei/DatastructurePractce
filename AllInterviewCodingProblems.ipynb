```python
from sys import argv

# Van Eck's sequence is a list of integers starting at zero, where
# the next number in the sequence is given by the distance between the 
# current number and the nearest preceding occurrence of that same number.
# ref.: http://blogs.perl.org/users/damian_conway/2019/07/as-simple-as-possiblebut-no-simpler.html

def van_eck(first=0):    
    index = {key: -1 for key in range(0, 10)}
    yield first

    last = first # last yielded element in the sequence
    last_index = 0
    while True: # yield the (last_index + 1)-th element
        try:
            curr = 0 if index[last] < 0 else last_index - index[last]
        except:
            index[last] = -1 # last wasn't on the dict yet
            curr = 0
        yield curr
        index[last] = last_index
        last = curr
        last_index += 1

try:
    n = int(argv[1])
    first = int(argv[2])
    g = van_eck(first)
    print([next(g) for _ in range(0, n)])
except:
    print('usage: python van-eck n [first]')
```

```python
def vanEckSequence(n):
    d=0
    next_num=0
    seq=""
    dct=dict()
    
    for i in range(n):
        if next_num in list(dct.keys()):
            d= i-dct[next_num]
        else:
            d=0
        dct[next_num]=i
        
        seq += str(next_num)
        next_num=d
        
    return ",".join(seq)
```

##### 2. Write a program to print twin primes less than 1000. If two consecutive odd numbers are both prime then they are known as twin primes

```python
def getPrimeTwins(n):
    
    primeTwins=[]
    
    for i in range(1,n+1):
        if i==1 or i==2:
            continue
            
        else:
            for j in range(2,int(i**0.5)+1):
                if i%j==0:
                    break
            else:
                primeTwins.append(i)
                
                    
    return [(primeTwins[i],primeTwins[i+1]) for i in range(len(primeTwins)-1)]
```

```python
getPrimeTwins(1000)
```

#### 4. Write a program to implement these formulae of permutations and combinations. Number of permutations of n objects taken r at a time: p(n, r) = n! / (n-r)!. Number of combinations of n objects taken r at a time is: c(n, r) = n! / (r!*(n-r)!) = p(n,r) / r!

```python
def permutation(n,r):
    def factorial(n):
        return 1 if n==1 else (n*factorial(n-1))
    return int(factorial(n)/factorial(n-r))
```

```python
permutation(4,3)
```

```python
def combination(n,r):
    def factorial(n):
        return 1 if n==1 else (n*factorial(n-1))
    return int(factorial(n)/(factorial(n-r)*factorial(r)))
```

```python
cube = lambda x: x**3;# complete the lambda function 

def fibonacci(n):
    # return a list of fibonacci numbers
    fibb=[0,1]
    if n>2:
        while len(fibb)<n:
            fibb.append(sum(fibb[::-1][0:2]))
        return fibb
    else:
        return fibb
        
        

if __name__ == '__main__':
    n = int(input())
    print(list(map(cube, fibonacci(n))))
```

```python
def isValid(s):
    opening = "({["
    closing = ")}]"
    #closing = "(}]"
    stack = []
    for i in s:
        if i in opening:
            print(i)
            stack.append(i)
            print(f"1 Here stack is {stack}")
        else:
            print(f"2 Here stack is {stack}")
            if len(stack) != 0  and stack[-1] == opening[closing.index(i)]:
                print(opening[closing.index(i)])
                stack.pop()
                print(f"3 Here stack is {stack}")
            else:
                return False

    return len(stack) == 0
```

```python
def is_subsequence(a, b):
    i = 0  # pointer for a
    for char in b:
        if i < len(a) and a[i] == char:
            i += 1
    return i == len(a)

# Example usage:
a = "abc"
b = "ahbgdc"
print(is_subsequence(a, b))  # Output: True

a = "acd"
b = "ahbgdc"
print(is_subsequence(a, b))  # Output: False
```
